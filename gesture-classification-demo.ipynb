# Data Augmentation

```python
# Randomly augment data. No need if we have enough data.
# fit for the first version dataset 
# may need some changes if we update the dataset
aug = 600
col=df.columns
df_aug=df.copy()

for i in range(aug):
    new_row = [np.concatenate((df_ori.sample(1).values[0,0:2],df_ori.sample(1).values[0,2:] + np.random.normal(0,5)),axis = 0)]
    df_gen = pd.DataFrame(new_row,columns = col)
    df_aug = pd.concat([df_aug,df_gen],ignore_index = True)
    
# Add more data whose letter is "A", to balance dataset
for i in range(600):
    new_row = [np.concatenate((df_ori.loc[df_ori['Letter']=="A",:].sample(1).values[0,0:2],df_ori.loc[df_ori['Letter']=="A",:].sample(1).values[0,2:] + np.random.normal(0,5)),axis = 0)]
    df_gen = pd.DataFrame(new_row,columns = col)
    df_aug = pd.concat([df_aug,df_gen],ignore_index = True)
    
# Add more data whose letter is "B", to balance dataset
for i in range(600):
    new_row = [np.concatenate((df_ori.loc[df_ori['Letter']=="B",:].sample(1).values[0,0:2],df_ori.loc[df_ori['Letter']=="B",:].sample(1).values[0,2:] + np.random.normal(0,5)),axis = 0)]
    df_gen = pd.DataFrame(new_row,columns = col)
    df_aug = pd.concat([df_aug,df_gen],ignore_index = True)
# Add more data whose letter is "C", to balance dataset
for i in range(600):
    new_row = [np.concatenate((df_ori.loc[df_ori['Letter']=="C",:].sample(1).values[0,0:2],df_ori.loc[df_ori['Letter']=="C",:].sample(1).values[0,2:] + np.random.normal(0,5)),axis = 0)]
    df_gen = pd.DataFrame(new_row,columns = col)
    df_aug = pd.concat([df_aug,df_gen],ignore_index = True)
```

# Machine Learning Model

```python
# Train the model, return the best model based on the test data
def train(train_X,train_y,test_X,test_y):
    lr = LogisticRegression(max_iter = 1000)
    lr.fit(train_X,train_y)
    lr_pre=lr.predict(test_X)
    lr_acc = metrics.accuracy_score(lr_pre,test_y)
    print('The accuracy of the Logistic Regression is',lr_acc)
    dt=DecisionTreeClassifier()
    dt.fit(train_X,train_y)
    dt_pre=dt.predict(test_X)
    dt_acc = metrics.accuracy_score(dt_pre,test_y)
    print('The accuracy of the Decision Tree is',dt_acc)
    rf = RandomForestClassifier(n_estimators=1000, criterion='entropy', random_state=None, bootstrap=True)
    rf.fit(train_X,train_y)
    rf_pre=rf.predict(test_X)
    rf_acc = metrics.accuracy_score(rf_pre,test_y)
    print('The accuracy of the Random Forest is',rf_acc)    
    knn=KNeighborsClassifier(n_neighbors=3) #this examines 3 neighbours for putting the new data into a class
    knn.fit(train_X,train_y)
    knn_pre=knn.predict(test_X)
    knn_acc = metrics.accuracy_score(knn_pre,test_y)
    print('The accuracy of the KNN is',knn_acc)    
    svm=SVC(probability=True)
    svm.fit(train_X, train_y)
    svm_pre=svm.predict(test_X)
    svm_acc = metrics.accuracy_score(svm_pre,test_y)
    print('The accuracy of the SVM is',svm_acc)
    mlp=MLPClassifier(hidden_layer_sizes=(128,64,32),random_state = 1)
    mlp.fit(train_X,train_y)
    mlp_pre=mlp.predict(test_X)
    mlp_acc = metrics.accuracy_score(mlp_pre,test_y)
    print('The accuracy of the MLP is',mlp_acc)
    dir = {lr:lr_acc,dt:dt_acc,rf:rf_acc,knn:knn_acc,svm:svm_acc,mlp:mlp_acc}
    best = 0
    model = None
    for i in dir.keys():
        if dir[i]>best:
            best = dir[i]
            model = i
    return model

    
```

```python
# train the model
model = train(train_X,train_y,test_X,test_y)
```

```python
def pred(model, X):
    prediction=model.predict_proba(X.reshape(1, -1))
    cl = model.classes_
    pro = prediction[0]
    head=""
    internal = ""
    for i in range(len(cl)):
        if i == 0:
            head = "{}:{}".format(cl[i],pro[i])
        else:
            internal = internal + (" ;{}:{}".format(cl[i],pro[i]))
        
    return "Probabilities: {" +head+internal+"}."
```

# Timestamp OCR from video

This script extracts timestamps from a video. You can change the video file path, limit the number of frames processed, and adjust the bounding box for the timestamp.

```python
from matplotlib import pyplot as plt
from cv2 import cv2
import pytesseract

input_video_path = 'c:/users/olli4/downloads/2022-03-18 17-07-41.mp4'  # Path to input video file
output_csv_path = '2022-03-18 17-07-41 timestamps.csv'  # Path to output CSV file where OCR'd timestamps will be written
start_frame = 0  # Start frame index (0-based)
max_num_frames = -1  # Maximum number of frames to process, or -1 to process the whole video
bbox = [[854, 1310], [908, 1916]]  # Bounding box [[top left y, top left x], [bottom right y, bottom right x]]

cap = cv2.VideoCapture(input_video_path)
fps = cap.get(cv2.CAP_PROP_FPS)
frame_index = 0
with open(output_csv_path, 'w') as output_file:
  while True:
    success, frame = cap.read()  
    if success:
      frame_grayscale = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
      frame_crop = frame_grayscale[bbox[0][0]:bbox[1][0], bbox[0][1]:bbox[1][1]]
      if frame_index == 0:
        print("Full first frame:")
        plt.imshow(frame_grayscale, cmap="gray")
        plt.show()        
        print("Cropped first frame:")
        plt.imshow(frame_crop, cmap="gray")
        plt.show()        
        cv2.imwrite('frame0.png', frame_grayscale)
        cv2.imwrite('frame0_crop.png', frame_crop)
        string = "frame_index,time,timestamp"
        print(string)
        print(string, file = output_file)
      if frame_index >= start_frame:
        timestamp = pytesseract.image_to_string(frame_crop, lang='eng', config='--psm 6 -c doc_dict_enable=0 -c load_system_dawg=0 -c load_freq_dawg=0 -c load_unambig_dawg=0 -c load_punc_dawg=0 -c load_number_dawg=0 -c load_fixed_length_dawgs=0 -c load_bigram_dawg=0 -c tessedit_char_whitelist="0123456789 -." -c numeric_punctuation=".-" -c chs_trailing_punct1="" -c language_model_penalty_punc=0 -c segment_penalty_dict_nonword=0 -c segment_penalty_garbage=0 -c language_model_penalty_non_freq_dict_word=0 -c language_model_penalty_non_dict_word=0')
        timestamp = timestamp.strip().replace('. ', '.') # Remove newline and fix some OCR errors
        string = str(frame_index) + ',' + str(frame_index/fps) + "," + timestamp
        print(string)
        print(string, file = output_file)
      frame_index += 1
      if max_num_frames != -1 and frame_index >= start_frame + max_num_frames:
        break;
    else:
      break
```
